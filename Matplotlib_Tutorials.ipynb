```python
#Draw multiple plots in one figure:
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])  
plt.subplot(1, 2, 1)                        # means the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)                         # means the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP", fontsize=20)        #TO add super title
plt.show()
```

```python
#If we want a figure with 2 rows an 1 column (meaning 2 plots will be displayed on top of each other instead of side-by-side).
#Use plt.subplot(2, 1, 1) & plt.subplot(2, 1, 2) instead of (1,2,1) & (1,2,2) in above syntax.
```

```python
#If we want to draw let's say 6 plot in 1 figure:
# Use plt.subplot(2, 3, 1) upto (2,3,6) OR plt.subplot(3,2,1) upto (3,2,6)
```

# Bar Chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.xlabel('Company')
plt.ylabel("Revenue(Bln)")
plt.bar(company, revenue, label="Revenue")
plt.title('US Technology Stocks')
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Multiple Bars showing revenue and profit of major US tech companies

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

# Horizontal bar chart using barh function

```python
plt.barh(xpos,revenue, label="Revenue") #we write .barh
#Note: For width of horizontal bars, use height instead of width.

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

# Pie chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels, radius=1.5)     #radius by default is 1.
plt.legend(title='Expenses')                        #Add legend with title
```

```python
#"equal" : to draw a perfect cicle pie chart, #plt.show: to remove those extra labels,
#add perrcentage,in %1.2f%% 2 means we want 2  digits after decimal
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.2f%%')
plt.show()
```

```python
#explode: in 0.4, 4 denotes the amount of explosion.
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.4])
plt.show()
```

```python
#change stasting angle. by defualt it starts from +ve X axis and goes in anti clockwise direction.
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2], startangle=45)
plt.show()
```

# For saving pie chart as .png

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.4])
plt.savefig("piechart.png", bbox_inches="tight")
```

```python
#pad_inches: if we want some extra space around chart in o/p png file
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.4])
plt.savefig("piechart.png", bbox_inches="tight", pad_inches=12)
plt.legend()
```
